### 1.3 Exceptions

Write `safe_divide(a, b)` that:
- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Inputs must be numeric")
    try:
        return a / b
    except ZeroDivisionError:
        return None

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None
try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"
print("1.3 passed ✓")
````

### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify A v = lambda v for each eigenpair
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    vec = eigenvectors[:, i]
    assert np.allclose(A @ vec, lam * vec)
    print(f"λ={lam:.2f}:  A@v = {np.round(A @ vec, 3)},  λ*v = {np.round(lam * vec, 3)}")

# Plot each eigenvector, stretched by its eigenvalue
fig, ax = plt.subplots()
colors = ['tab:red', 'tab:green']
for i in range(len(eigenvalues)):
    scaled = eigenvectors[:, i] * eigenvalues[i]
    ax.quiver(0, 0, scaled[0], scaled[1], angles='xy', scale_units='xy', scale=1,
              color=colors[i], width=0.01, label=f'λ={eigenvalues[i]:.1f}')
ax.set_xlim(-6, 6); ax.set_ylim(-6, 6)
ax.axhline(0, color='k', lw=0.8); ax.axvline(0, color='k', lw=0.8)
ax.set_aspect('equal')
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_title('Eigenvectors scaled by eigenvalues')
ax.legend()
plt.tight_layout(); plt.show()

print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

When you multiply an ordinary vector by a matrix it usually changes both its length and its direction. An eigenvector is the special case where the direction stays the same and only the length changes. The eigenvalue is the scaling factor: if |λ| > 1 the vector gets stretched, if 0 < |λ| < 1 it gets shrunk towards the origin, and if λ is negative the vector also flips to point the other way. For this matrix the eigenvalue 5 is larger than 2, so the vector lying along that eigenvector is stretched the most. That is why its arrow comes out longest in the plot, while the λ = 2 direction is stretched much less.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)

X_reconstructed = U @ np.diag(S) @ Vt
X_approx        = S[0] * np.outer(U[:, 0], Vt[0, :])   # keep only the top component

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

SVD splits X into three parts, U, diag(S) and Vᵀ. The rows of Vᵀ are the right singular vectors and they act like a fresh set of perpendicular axes for the data. The values in S say how far the data spreads out along each of those axes, and U gives the coordinates of every original row in that new axis system. If you mean-centre X first (subtract each column mean), those rows of Vᵀ are the same as the principal components you get from PCA, and each singular value squared is proportional to the variance that component explains. So the directions of maximum variance are the ones stored in Vᵀ, and the direction tied to the biggest singular value is the one the data varies along the most. Dimensionality reduction is just keeping the top few of these and throwing the rest away.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each sample against the standard normal CDF
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")
# s1 keeps a high p (can't reject normal); s2's p is ~0 so it is clearly not normal.

# ADF test on a series with an obvious upward trend
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")
# p > 0.05 -> fail to reject the unit-root null -> the trended series is non-stationary.

# Difference once to strip the trend, then re-test
ts_diff = np.diff(ts)
adf_diff = adfuller(ts_diff)
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    eps = 1e-10
    # bucket edges come from the reference (expected) distribution's quantiles
    edges = np.percentile(expected, np.linspace(0, 100, bins + 1))
    edges[0], edges[-1] = -np.inf, np.inf      # let the end bins catch the tails
    exp_pct = np.histogram(expected, bins=edges)[0] / len(expected) + eps
    act_pct = np.histogram(actual,   bins=edges)[0] / len(actual)   + eps
    return np.sum((act_pct - exp_pct) * np.log(act_pct / exp_pct))

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
ax.hist(train_dist, bins=30, density=True, alpha=0.6, color='tab:blue',   label='Training')
ax.hist(drift_dist, bins=30, density=True, alpha=0.6, color='tab:orange', label='Production')
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** the relationship between the inputs and the target changes, so P(y | X) shifts and the same input should now give a different output. For example, what counted as fraud last year might look normal this year.
- **Covariate drift:** only the input distribution P(X) changes while the rule linking inputs to the target stays valid. For example our users suddenly skew younger, but age still affects the target the same way.
- **PSI < 0.1:** the distribution is stable, no action needed.
- **PSI 0.1 to 0.2:** a small to moderate shift, worth keeping an eye on.
- **PSI > 0.2:** a big shift, the production data has moved a long way from the training data. Our example PSI of 1.54 sits here.
- **Retraining trigger example:** if the PSI of an important feature stays above 0.2 for several days in a row, or a live metric like accuracy or AUC drops below an agreed level, that is a good signal to retrain on newer data.

## Part 6 — Probability Theory

### 6.1 Core Concepts

A bag contains: 4 red, 3 blue, 3 green marbles.

1. Define the sample space and compute P(red), P(blue), P(green).
2. Two marbles drawn **without replacement**. Compute:
   - P(first=red, second=blue) — **joint probability**
   - P(second=blue | first=red) — **conditional probability**
3. Are the two draws **independent**? Show mathematically.

````python
total = 10
red, blue, green = 4, 3, 3

p_red   = red / total
p_blue  = blue / total
p_green = green / total

# Joint: P(red first AND blue second) = P(red) * P(blue | red)
p_red_then_blue = (red / total) * (blue / (total - 1))

# Conditional: one red is already gone, so 9 marbles remain
p_blue_given_red = blue / (total - 1)

# Draws are independent only if P(B | A) == P(B)
independent = bool(np.isclose(p_blue_given_red, p_blue))

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue):     {p_blue:.4f}")
print(f"Independent? {independent}")
````

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) via the law of total probability
p_free = p_free_given_spam * p_spam + p_free_given_ham * p_ham

# Bayes' theorem
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    prior_ham = 1 - prior_spam
    evidence  = p_word_given_spam * prior_spam + p_word_given_ham * prior_ham
    return (p_word_given_spam * prior_spam) / evidence

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | 0.30, the chance any email is spam before we read it |
| Likelihood | P(word \| Spam) | 0.80, the chance a spam email contains the word "free" |
| Evidence | P(word) | 0.275, the chance that any email at all contains "free" |
| Posterior | P(Spam \| word) | 0.873, the updated belief it is spam once we see "free" |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means = np.array([
    np.random.choice(population, size=sample_size).mean()
    for _ in range(n_samples)
])

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std / np.sqrt(sample_size)   # CLT: std of the sample mean

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot histogram of the sample means with the CLT normal on top
fig, ax = plt.subplots()
ax.hist(sample_means, bins=40, density=True, alpha=0.6, color='tab:blue',
        edgecolor='white', label='Sample means')
xs = np.linspace(sample_means.min(), sample_means.max(), 200)
ax.plot(xs, stats.norm.pdf(xs, pop_mean, clt_std), 'r-', lw=2, label='CLT normal')
ax.legend()
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

The population here is a strongly right-skewed exponential, but the histogram of the sample means still turns into a roughly symmetric bell shape centred on the population mean of about 1.0. Its spread also lands very close to the CLT formula σ/√n (about 0.183 either way), which is what the theorem says should happen. The KS test still gives a tiny p-value, but that is mainly because there are 5,000 sample means, which is enough for the test to notice the small skew that is still left when you only average 30 exponential values. That leftover skew keeps shrinking as the sample size grows. This is why CLT is so useful: it lets us use normal-based methods such as confidence intervals and t-tests on the mean of almost any data, and that approximate normality is the assumption behind a lot of the classic statistical tests.
